```python
#importando as bibliotecas necessárias
import matplotlib.pyplot as plt
from pywaffle import Waffle

fig = plt.figure(
    FigureClass=Waffle,
    rows=6,
    title={'label': 'Para qual time de futebol você torce?', 'loc': 'left'},
    values={ 
        'Palmeiras': 31,
        'Corinthians': 40,
        'São Paulo': 37,
        'Santos': 18
    },
    legend={'loc': 'upper left',
            'bbox_to_anchor': (1.1, 1)            
    },
    figsize=(10,10)
)

fig.set_tight_layout(False)
plt.show()
```

##Customização do gráfico de waffle

```python
#Importar o pandas
import pandas as pd
#ler um arquivo CSV obtido do site Inside Airbnb, com inforações de acomodações oferecidos dentro do Airbnb para Barcelona
df = pd.read_csv("http://data.insideairbnb.com/spain/catalonia/barcelona/2022-09-10/visualisations/listings.csv",quotechar='"')

#extrair as quantidades de cada um dos tipos de imóvel p/ alugar no Airbnb
data = df['room_type'].value_counts().to_dict()
# plotar gráfico de waffle
fig = plt.figure(
                title={
                    'label': 'Tipos de acomodação oferecidos pelo Airbnb na cidade de Barcelona',
                    'loc': 'left',
                            'fontdict': {
                                'fontsize': 20
                            }
                },
                FigureClass=Waffle,
                rows=5,
                columns=20,
                colors=("#232066", "#983D3D", "#DCB732",'#1B7C2A'),
                values=data, icons='bed', font_size=20,
                legend={'loc': 'lower left', 'fontsize':14,  'bbox_to_anchor': (0, -0.2), 'ncol': len(data), 'framealpha': 0},
                labels=[f"{k} ({(v / sum(data.values()) * 100):.2f}%)" for k, v in data.items()], 
                icon_legend=True,
                figsize=(20,10)
                )
fig.set_tight_layout(False)
plt.show();
```
